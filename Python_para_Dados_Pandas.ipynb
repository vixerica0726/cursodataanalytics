```python
# Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros

# Contar a quantidade de carros por estado
estado_counts = df['Estado'].value_counts()
# Exibir a contagem de carros por estado, ordenado de forma decrescente
print(estado_counts)

# Contar a quantidade de carros por estado e adicionar como nova coluna no dataframe
df['quantidade_carros'] = df.groupby('Estado')['Estado'].transform('count')

# Ordenar o dataframe de forma decrescente pela quantidade de carros
df_sorted = df.sort_values(by='quantidade_carros', ascending=False)

# Mostrar o dataframe ordenado
df_sorted.head(10)
```

```python
# Faça um dataframe com  Estado e proporção de carro por habitante

# Calcular a proporção de carros por habitante
df['Proporcao_Carros_Habitante'] = df['Quant Carros'] / df['População do Estado']

# Criar um DataFrame apenas com Estado e a proporção de carros por habitante
df_proporcao = df[['Estado', 'Proporcao_Carros_Habitante']]

# Exibir o novo DataFrame
print(df_proporcao)
```

# Agregação de dados

**Utilizando o arquivos: concurso.csv**
7. Qual a média de idade por região?
8. Qual é a porcentagem de pessoas com deficiência?
9. Quantas pessoas são do estado do Piauí?


```python
from google.colab import files
uploaded = files.upload() #upload

import pandas as pd
from datetime import datetime

# Carregar o arquivo CSV
df = pd.read_csv('dados_concurso.csv')

# Verificar se a coluna 'Data de Nascimento' está no formato correto
df['Data de Nascimento'] = pd.to_datetime(df['Data de Nascimento'], errors='coerce')

# Calcular a idade a partir da data de nascimento
df['Idade'] = df['Data de Nascimento'].apply(lambda x: (datetime.now() - x).days // 365 if pd.notnull(x) else None)

# Média de idade por estado (já que não temos a coluna 'Região')
media_idade_por_estado = df.groupby('Estado')['Idade'].mean()
print("Média de idade por estado:")
print(media_idade_por_estado)

# Porcentagem de pessoas com deficiência
porcentagem_deficiencia = df['Deficiência'].value_counts(normalize=True).get('Sim', 0) * 100
print(f'\nPorcentagem de pessoas com deficiência: {porcentagem_deficiencia}%')

# Número de pessoas do estado do Piauí
pessoas_piaui = df[df['Estado'] == 'PI'].shape[0]
print(f'\nNúmero de pessoas do estado do Piauí: {pessoas_piaui}')
```

# Valores faltosos

14. Para os valores faltosos coloque:
- Se string: Falta informação
- Se número: mediana
15. Plote a porcentagem de dados faltosos para cada coluna.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
df = pd.read_csv('dados_concurso.csv')

# Verificando valores faltosos
faltosos = df.isnull().mean() * 100  # Porcentagem de valores faltosos em cada coluna

# Substituir valores faltosos:
for col in df.columns:
    if df[col].dtype == 'object':  # Se a coluna for de tipo string
        df[col].fillna('Falta informação', inplace=True)
    else:  # Se a coluna for numérica
        df[col].fillna(df[col].median(), inplace=True)

# Plotando a porcentagem de dados faltosos
plt.figure(figsize=(10, 6))
faltosos.plot(kind='bar', color='skyblue')
plt.title('Porcentagem de Dados Faltosos por Coluna')
plt.ylabel('Porcentagem')
plt.xlabel('Colunas')
plt.xticks(rotation=45)
plt.show()

# Exibindo o DataFrame após a substituição
print(df.head())  # Exibindo as primeiras linhas para verificar as substituições
```
